```python
import random as rd
```

```python
def obtener_edad(d):
    edad=str(rd.randint(30,40))
    edad=edad+" años"
    return edad
```

```python
provincias=['Loja', 'Pichincha', 'Guayas', 'Azuay', 'Manabí']
```

```python
def obtener_provincia(p):
    provincia=rd.choice(provincias)
    return provincia
```
